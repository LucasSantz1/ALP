### 1. Faça um programa que leia uma matriz 2x3 (2 linhas, 3 colunas). Apresente os elementos da matriz e seus respectivos índices.

```
numeros = []
for linha in range (2):
    vet_linha = []
    for coluna in range (3):
        vet_linha.append(int(input(f'Digite o valor [{linha}][{coluna}]:')))
    numeros.append(vet_linha)
print('\nApresentado a matriz de forma simples')
print(numeros)
print('\nApresentado a matriz de forma tabular')
for linha in range(len(numeros)):
    for coluna in range(len(numeros[0])):
        print(numeros[linha][coluna], end = '\t')

for linha in range(len(numeros)):
    for coluna in range(len(numeros[0])):
        print(f'Elemento: {numeros[linha][coluna]}\tÍndice da linha: {linha} \tÍndice da coluna: {coluna}')

```

### 2. Faça um programa que carregue uma matriz 2 x 2, que representa o salário de 4 funcionários, calcule e mostre a soma total de todos os elementos que será o montante pago pela empresa a esses funcionários.

```
salario = []
soma = 0
for linha in range(2):
    vetlinha = []
    for coluna in range(2):
        vetlinha.append(float(input(f'Digite o solario[{linha}][{coluna}] R$: ')))
        soma = soma + vetlinha[coluna]
    salario.append(vetlinha)
print('Na soma dos salarios que a empresa pagará será de R$:',soma  )
```

### 3. Faça um programa que carregue uma matriz 3 x 2, que representa preços de produtos, crie OUTRA matriz que armazene todos os preços com 7% de aumento.

```
preco = []
aumento = []
for linha in range(3):
    vet_linha = []
    for coluna in range(2):
        vet_linha.append(float(input(f'Digite o preço[{linha}][{coluna}] R$: ')))
    preco.append(vet_linha)    

print('\nApresentação da matriz aumento........')
for linha in range(len(preco)):
    vet_linha = []
    for coluna in range(len(preco[0])):
        vet_linha.append(preco[linha][coluna]*7/100+preco[linha][coluna])
        print(vet_linha[coluna],end='\t')
    aumento.append(vet_linha)
    print()
```

### 4. Faça um programa para armazenar em uma matriz 5x2 preços. Encontre e apresente os ÍNDICES dos valores menores que 23 reais.

```
preco = []
for linha in range(5):
    vetlinha = []
    for coluna in range(2):
        vetlinha.append(float(input(f'Digite o preco [{linha}][{coluna}]')))
    preco.append(vetlinha)
for linha in range(len(preco)):
    for coluna in range(len(preco[0])):
        if preco[linha][coluna] < 23:
            print(f'linha:{linha} coluna:{coluna}')
```

### 5. Faça um programa que leia números inteiros m e n e os elementos de uma matriz A de números inteiros de dimensão m x n e conte o número de elementos que são iguais a zero.


```
mtz = []
m = int(input('Informe a quantidade de linhas da matriz'))
n = int(input('Informe a quantidade de colunas da matriz'))
i = 0
for linha in range(m):
    v = []
    for coluna in range(n):
        v.append(int(input(f'Digite um numero[{linha}][{coluna}]:')))
    mtz.append(v)
    if v [coluna] == 0:
        i += 1
print(mtz)
print(f'A quantidade de elementos iguais a zero é: {i}')


```
