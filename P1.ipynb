###**1**. [1,50 pontos] Você está em uma festa de aniversário e está tendo confusão para distribuir de modo justo a quantidade de brigadeiros entre os participantes.

###Todos, por alguma coincidência, entendem de Python, então você planeja o seguinte exemplo de código:

* brigadeiros = 150

* pessoas = 16

* Por você ter feito a conta e evitado aumento da confusão, todos decidiram dar a você além da divisão justa, o resto dos brigadeiros que sobraram da conta.

###Crie um algoritmo utilizando a linguagem Python para calcular e apresentar a sua quantidade de brigadeiros. Adote os nomes das variáveis apresentadas para o desenvolvimento e siga o planejamento.


```
brigadeiros = int(input('Digite a quantidade de brigadeiros: '))
pessoas = int(input('Digite a quantidade de pessoas na festa: '))
sobra = int(input('Digite a quantidade que sobrou: '))
quantidade = (brigadeiros - sobra) / pessoas
eu = quantidade + sobra
print(f'A divisão igual de brigadeiros é: {quantidade:.0f}')
print(f'A quantidade de brigadeiros recebidos por mim é: {eu:.0f}')

```

###[2,00 pontos] **2.** Um professor de uma instituição de ensino, pediu ajuda para você organizar as notas dos alunos de três provas, que ocorreram ao longo do semestre, mas ainda a última não foi aplicada. Ele informou que para o aluno ser aprovado é necessário ter a média das notas igual a 7.
###O professor passou algumas diretrizes na qual você converteu em Python:
* media = 7

* numero_provas = 3

* O aluno é considerado **reprovado de forma direta**, se não realizar uma das provas.

###Crie um programa utilizando a linguagem Python, para calcular a média e avaliar a situação de aprovação ou não do aluno. 

```
prova1 = float(input('Digite a nota da primeira prova: '))
prova2 = float(input('Digite a nota da segunda prova: '))
prova3 = float(input('Digite a nota da terceira prova: '))
media = (prova1 + prova2 + prova3) / 3
if media >=7 and media <= 10:
    print('Aprovado')
    print(f'A média é {media:.1f}')
elif prova1 == 0 and prova2 == 0 and prova3 == 0:
     print('Reprovado')
     print(f'A média é {media:.1f}')
else:
    print('Reprovado')
    print(f'A média é {media:.1f}')
```

###Atendendo a 10 proprietários de automóveis, a partir dos anos de fabricação e dos valores dos carros, crie um programa utilizando a linguagem Python que, calcule e apresente:
a) Quantidade de carros isentos

b) Valor do IPVA de cada carro

c) Média de valores dos carros

d) Total dos IPVAs

```
carros = 0
isento = 0
nao_isento = 0
preco = 0
soma_ipva = 0
while carros < 3:
    idade = int(input('Digite o ano de fabricação do seu carro: '))
    valor = float(input('Digite o valor do seu carro: '))
    if idade < 2002:
        isento += 1
        carros += 1
        preco = preco + valor
        print('O seu carro é isento de IPVA')
        
    else:
        ipva = valor * (4/100)
        print('O ipva do seu carro é:',ipva)
        carros += 1
        nao_isento += 1
        preco = preco + valor
        soma_ipva = soma_ipva + ipva

media = preco / carros



print('A quantidade carros isentos é:',isento)
print(f'A Média de valores dos carros é: {media:.0f}')
print('O total dos IPVAs é: ',soma_ipva)




    
```
