### 1. Faça um programa que calcule e apresente a média de idades de uma sala de 35 alunos.​​

```
idade = [] 
soma_idade = 0
for i in range(3):
    idade.append(int(input('Informe a '+ str(i+1) +'ª idade: ')))
    soma_idade = soma_idade + idade[i]
media = soma_idade / len(idade)
print(f'A média é {media:.2f} anos.')
```

### 9. Criar um algoritmo que leia dados para um vetor de 100 elementos inteiros, imprimir o maior, o menor, o percentual de números pares e a média dos elementos do vetor. Obs.: percentual = quantidade contada * 100 / quantidade total. Não use nenhuma função pronta da linguagem Python.


```
num = []
soma_pares = 0
soma_elementos = 0
for i in range (5):
    num.append(int(input('Digite um número inteiro: ')))
    if i == 0:
        maior = num[i]
        menor = num[i]
    elif num[i] >= maior:
        maior = num[i]
    elif num[i] <= menor:
        menor = num[i]

for i in range(len(num)):
    if num [i] % 2 == 0:
        soma_pares += 1
        percentual = soma_pares * 100 / len(num)

for i in range(len(num)):
    soma_elementos = soma_elementos + num[i]
media = soma_elementos / len(num)

print('O maior número é',maior)
print('O menor numero é',menor)
print('O percentual de números pares é ',percentual,'%')
print('A média de elementos é:',media)



```
